```python
import torch
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F

import matplotlib.pyplot as plt

from tqdm import tqdm

import fairness_metrics
```

```python
class Trainer:
    def __init__(self, model, accloss, fairloss, N, Na, tester, regularizer=2, lr=1e-2):
        self.model = model
        self.optimizer = optim.Adam(model.parameters(), lr=lr)
        self.accloss = accloss
        self.fairloss = fairloss
        self.tester = tester
        self.N = N
        self.Na = Na
        self.X = None
        self.y = None
        self.a = None
        self.regularizer = regularizer
        self.tester(self.model)
    
    def update(self, X, y, a):
        '''
        Perform model update
        X, y, a: torch.Tensor
        
        '''
        # update X, y, a
        if self.X == None:
            self.X = X
            self.y = y
            self.a = a
        else:
            self.X = torch.vstack((self.X, X))
            self.y = torch.hstack((self.y, y))
            self.a = torch.hstack((self.a, a))
            
        # update if enough data
        if (len(self.a) >= self.N) and ((1 - self.a).sum() >= self.Na[0]) and (self.a.sum() >= self.Na[1]):
            # perform training step
            self.optimizer.zero_grad()
            y_hat = self.model(self.X)
            y_hat_1 = y_hat[self.a==1]
            y_hat_0 = y_hat[self.a==0]
            loss = self.accloss(y_hat, self.y) + self.regularizer * self.fairloss(y_hat_1, y_hat_0)
            loss.backward()
            self.optimizer.step()
            
            # reset data
            self.X = None
            self.y = None
            self.a = None
        
        # perform tests
        self.tester(self.model)
```

```python
class Tester:
    def __init__(self, X, y, a):
        self.X = X
        self.y = y
        self.a = a
        
        self.MSEs = []
        self.SPDs = []
        
    def test(self, model):
        y_hat = model(self.X)
        MSE = ((self.y.flatten()-y_hat.flatten())**2).mean()
        
        SPD = torch.tensor(0)
        y_hat_1 = y_hat[self.a==1]
        y_hat_0 = y_hat[self.a==0]
        for y_test in torch.hstack((y_hat_1.flatten(),y_hat_0.flatten())).flatten():
            cdf1_y = (y_hat_1<=y_test).float().mean()
            cdf2_y = (y_hat_0<=y_test).float().mean()
            if (cdf1_y-cdf2_y).abs()>SPD:
                SPD = (cdf1_y-cdf2_y).abs()
                
        self.MSEs.append(MSE.detach().numpy()[()])
        self.SPDs.append(SPD.detach().numpy()[()])
        
    def __call__(self, model):
        self.test(model)
```

```python
k = 10
N_iter = 2000
```

```python
slopes = 4*torch.rand((k, 5))-2
```

```python
def get_sample(k):
    X = torch.rand((1,k-1))
    a = torch.randint(2,(1,))
    X = torch.hstack((X,torch.unsqueeze(a, 1)))
    y = (X @ slopes).max()
    return X, y, a
```

```python
def build_test_set(k, N_test=1000):
    X = torch.rand((N_test,k-1))
    a = torch.randint(2,(N_test,))
    X = torch.hstack((X,torch.unsqueeze(a, 1)))
    y = (X @ slopes).max(dim=1)[0]
    return X, y, a
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self, k):
        super(NeuralNetwork, self).__init__()
        self.linear1 = torch.nn.Linear(k, 20, bias=True)
        self.linear2 = torch.nn.Linear(20, 1, bias=True)

    def forward(self, x):
        x = F.relu(self.linear1(x))
        self.output = self.linear2(x)
        return self.output
```

```python
def mse(yhat, y):
    return ((yhat.flatten()-y.flatten())**2).mean()
```

```python
def test(N=4, seed=0):
    test_set = build_test_set(k)
    tester_ED = Tester(*test_set)
    tester_WD = Tester(*test_set)
    tester_WD_bign = Tester(*test_set)

    torch.manual_seed(seed)
    trainer_ED = Trainer(NeuralNetwork(k), mse, fairness_metrics.energy_distance, N, [2,2], tester_ED)
    torch.manual_seed(seed)
    trainer_WD = Trainer(NeuralNetwork(k), mse, fairness_metrics.sinkhorn_diver, N, [2,2], tester_WD)
    torch.manual_seed(seed)
    trainer_WD_bigN = Trainer(NeuralNetwork(k), mse, fairness_metrics.sinkhorn_diver, 50, [2,2], tester_WD_bign)

    for i in tqdm(range(N_iter)):
        sample = get_sample(k)
        trainer_ED.update(*sample)
        trainer_WD.update(*sample)
        trainer_WD_bigN.update(*sample)

    return (trainer_ED.tester.MSEs,
           trainer_ED.tester.SPDs,
           trainer_WD.tester.MSEs,
            trainer_WD.tester.SPDs,
           trainer_WD_bigN.tester.MSEs,
            trainer_WD_bigN.tester.SPDs,)
```

```python

MSE_EDs,SPD_EDs,MSE_WDs,SPD_WDs, MSE_WD_bigNs,SPD_WD_bigNs = [],[],[],[],[],[]
for i in range(20):
    MSE_ED,SPD_ED,MSE_WD,SPD_WD, MSE_WD_bigN,SPD_WD_bigN=test(seed=i)
    MSE_EDs.append(MSE_ED)
    SPD_EDs.append(SPD_ED)
    MSE_WDs.append(MSE_WD)
    SPD_WDs.append(SPD_WD)
    MSE_WD_bigNs.append(MSE_WD_bigN)
    SPD_WD_bigNs.append(SPD_WD_bigN)
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.melt(pd.DataFrame(MSE_EDs).T.reset_index(), id_vars='index')
df.columns = ['Sample', 'b', 'MSE']
sns.lineplot(data = df, 
             x="Sample", y="MSE", label='ED')

df = pd.melt(pd.DataFrame(MSE_WDs).T.reset_index(), id_vars='index')
df.columns = ['Sample', 'b', 'MSE']
sns.lineplot(data = df, 
             x="Sample", y="MSE", label='WD')

df = pd.melt(pd.DataFrame(MSE_WD_bigNs).T.reset_index(), id_vars='index')
df.columns = ['Sample', 'b', 'MSE']
sns.lineplot(data = df, 
             x="Sample", y="MSE", label='WD-bigN')
plt.yscale('log')
```
